# What we should do:
#### 1) renaming columns for easier usage
#### 2) converting wrong data types like for column `AGE`
#### 3) imputing `nan` values, or dropping them if they aren't important

```python
# Renaming Columns

new_col=[]
for col in df.columns:
    x = col.find(":")
    col = col[x+1:].strip()
    new_col.append(col)
df.columns = new_col

df.rename(columns={"STATE, PROVINCE, COUNTY, ETC" : "AREA"},inplace =True)

df.columns
```

```python
# converting 'AGE' from object to int

df["AGE"].fillna("0",inplace = True)

for record in df["AGE"]:
    if record.isdigit():
      pass
    else:
        df.loc[(df["AGE"]==record), 'AGE'] = "41"
       
df["AGE"]= pd.to_numeric(df["AGE"])
```

```python
df.loc[(df["AGE"] > 90) | (df["AGE"]< 7 ), "AGE"] = 41
```

```python
df["AGE"].unique()
df.columns
```

```python
pd.set_option('display.max_rows',None)
df[df.columns].isnull().sum()
```

```python
# dropping unnecessary columns

df = df.drop(["Internal ID","AREA","JOY OTHER","DESPAIR OTHER","OTHER COMMENTS","MEDIA [Daily Dish]"\
             ,"MEDIA [Science]","MEDIA [ESPN]","MEDIA [Yahoo]","113",'Click Coordinates (x, y)'],axis=1)    
    
# 'Internal ID' has no benfit
# 'AREA' we can benefit from 'Country' instead
# they have too many nan_values
# 'Click Coordinates (x, y)' I don't understand the intution behind its usage
```

```python
df[df.columns].isnull().sum()
```

```python
# imputing the nan values in "Q6" columns with the most frequent value in each column

for col in df.columns[4:108]:
    md = df[col].mode().iat[0]
    df[col] = df[col].fillna(md)
```

```python
# imputing nan values for "DRESS" column

df["DRESS"] = df["DRESS"].astype("category")
df["DRESS"].value_counts()
df["DRESS"] = df["DRESS"].fillna("White and gold")
```

```python
df[df.columns].isnull().sum()
```

```python
# Dealing with the different enteries of the same Country in "COUNTRY" column

df["COUNTRY"] = df["COUNTRY"].str.lower()

df["COUNTRY"].value_counts()

from fuzzywuzzy import process

Right_countries = ["atlantis","australia","california","canada","china","costarica","denemark"\
                  ,"england","germany","ireland","japan","korea","mexico"\
                  , "netherlands","singapore","sweden","switzerland","usa","united states"\
                  ,"uk","united kingdom"]


for country in Right_countries:
    matches = process.extract(country,df["COUNTRY"],limit = df.shape[0])
    for potential_match in matches:
        if potential_match[1] >= 70:
            df.loc[df["COUNTRY"] == potential_match[0] , "COUNTRY"] = country
```

```python
 df['COUNTRY'].unique()
```

```python
df["COUNTRY"]= df["COUNTRY"].replace(['35', '46', '45', '32', '1',"subscribe to dm4uz3 on youtube","narnia","murica","murrika"\
                                       ,"i don't know anymore","fear and loathing",'soviet canuckistan','insanity lately'\
                                       ,'i pretend to be from canada, but i am really from the united states.'\
                                       ,'north carolina ','ud'], np.nan)


for i in ['united states','america','usa? hard to tell anymore..',"'merica","ahem....amerca",'n. america'\
         ,'california','new jersey','new york']:
    df["COUNTRY"] = df["COUNTRY"].replace({i:"usa"})

    
df["COUNTRY"] = df["COUNTRY"].replace({'pittsburgh':"russia","united kingdom":"uk"})


df["COUNTRY"] = df["COUNTRY"].replace({np.nan:"usa"})   # after running----> df["COUNTRY"].mode()
```
